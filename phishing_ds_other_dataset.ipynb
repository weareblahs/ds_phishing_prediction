## Additional data processing to match ultra safe features to ensure more accuracy

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_phishing.csv')
df2 = df.copy()
df3 = pd.DataFrame()
```

```python
def extract_domain(dom):
    d = str(dom)
    return d.replace("http://", "").replace("https://", "").split("/")[0]
df2['url'] = df.url.apply(extract_domain)
```

```python
df3
```

```python
ultra_safe_features = [
    'DomainLength',
    'TLD',
    'TLDLength',
    'NoOfSubDomain',
    'LetterRatioInURL',
    'label'
]
```

```python
df3['DomainLength'] = df2.url.apply(lambda x: len(x))
```

```python
df3['TLD'] = df2.url.apply(lambda x: x.split(".")[-1])
```

```python
df3['TLDLength'] = df3.TLD.apply(lambda x: len(x))
```

```python
df3['NoOfSubDomain'] = df2.url.apply(lambda x: len(x.split(".")) - 2)
```

```python
import re
from difflib import SequenceMatcher
df3['LetterRatioInURL'] = df2.url.apply(lambda x: SequenceMatcher(a="".join(re.findall(r'[a-zA-Z]+', x)), b=x).ratio())
```

```python
df3['label'] = df2.status.apply(lambda x: 0 if (x == "phishing") else 1)
```
